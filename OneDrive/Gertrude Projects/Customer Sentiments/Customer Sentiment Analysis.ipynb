```python
def to_long(df, col1, col2):

    grouped_df = df[[col1, col2]].groupby([col1, col2]).value_counts().to_frame().reset_index()
    new_df = pd.DataFrame(sorted(df[col1].unique()), columns=[col1])
    new_cols = df[col2].unique()
    
    for col in new_cols:
        new_df = new_df.merge(grouped_df[grouped_df[col2]==col], how='left')
        new_df.drop(columns=col2, inplace=True)
        new_df.rename(columns={'count':col}, inplace=True)
        new_df.replace(np.nan, 0, inplace=True)

    return new_df
```

### 1. Rank products by most positive %

```python
pos_percent = to_long(data, 'Source', 'Sentiment')
pos_percent['%_Positive'] = (pos_percent['Positive'] / (pos_percent['Positive'] + pos_percent['Negative']))#.transform(lambda x: f'{x: 0.00%}')
pos_percent.sort_values(by=['%_Positive', 'Source'], ascending=[False, True], inplace=True)
```

```python
all_pos = pos_percent[pos_percent['%_Positive']==1]
all_pos['%_Positive'] = all_pos['%_Positive'].transform(lambda x : f'{x:0.0%}')
all_pos
```

```python
all_neg = pos_percent[pos_percent['%_Positive']==0]
all_neg['%_Positive'] = all_neg['%_Positive'].transform(lambda x : f'{x:0.0%}')
all_neg
```

```python
inbtw = pos_percent[pos_percent['%_Positive']>0][pos_percent['%_Positive']<1]
inbtw['%_Positive'] = inbtw['%_Positive'].transform(lambda x : f'{x:0.0%}')
inbtw
```
